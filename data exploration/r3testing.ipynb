```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inp_df0 = pd.read_csv("../data/prices_round_3_day_0.csv",delimiter=";")
inp_df0
```

```python
inp_df1 = pd.read_csv("../data/prices_round_3_day_1.csv",delimiter=";")
inp_df1
```

```python
inp_df2 = pd.read_csv("../data/prices_round_3_day_2.csv",delimiter=";")
inp_df2
```

```python
master = pd.concat([inp_df0,inp_df1,inp_df2], axis = 0, ignore_index=True)
master
```

```python
df1 = pd.DataFrame(master[master["product"] == "CHOCOLATE"]["mid_price"]).reset_index(drop=True)
df2 = pd.DataFrame(master[master["product"] == "ROSES"]["mid_price"]).reset_index(drop=True)
df3 = pd.DataFrame(master[master["product"] == "GIFT_BASKET"]["mid_price"]).reset_index(drop=True)
df4 = pd.DataFrame(master[master["product"] == "STRAWBERRIES"]["mid_price"]).reset_index(drop=True)
```

```python
df_prices = pd.concat([df1,df2,df3,df4],axis=1)
df_prices = df_prices.set_axis(["choc","rose","basket","strawb"],axis=1)
df_prices["enter_ub"] = 50
df_prices["enter_lb"] = -1*df_prices["enter_ub"]
df_prices["exit_ub"] = 5
df_prices["exit_lb"] = -1*df_prices["exit_ub"]
df_prices["extreme_ub"] = 120
df_prices["extreme_lb"] = -120
```

```python
df_prices
```

```python
df_prices["debasket"] = df_prices["choc"]*4 + df_prices["rose"]*1 + df_prices["strawb"]*6
df_prices["diffs"] = df_prices["basket"]-(df_prices["debasket"])
df_prices
```

```python
print(df_prices["diffs"].mean())
```

```python
print((df_prices["basket"]-df_prices["debasket"]-382).mean())
```

```python
df_prices["timestamp"] = df_prices.index*100
```

```python
import matplotlib.pyplot as plt
import numpy as np

enter_trade = 50
exit_trade = 15

# Assuming your DataFrame is named df_prices
fig, axs = plt.subplots(4, 1, figsize=(12, 12), sharex=True)

# Plot choc
axs[0].plot(df_prices.index, df_prices['choc'], label='choc', color='b')
axs[0].set_ylabel('choc', color='b')
axs[0].tick_params(axis='y', labelcolor='b')

# Plot rose
axs[1].plot(df_prices.index, df_prices['rose'], label='rose', color='g')
axs[1].set_ylabel('rose', color='g')
axs[1].tick_params(axis='y', labelcolor='g')

# Plot strawb
axs[2].plot(df_prices.index, df_prices['strawb'], label='strawb', color='r')
axs[2].set_ylabel('strawb', color='r')
axs[2].tick_params(axis='y', labelcolor='r')

# Plot basket
axs[3].plot(df_prices.index, df_prices['basket'], label='basket', color='c')
axs[3].set_ylabel('basket', color='c')
axs[3].tick_params(axis='y', labelcolor='c')

# Mark points where diffs >= 60
for ax in axs:
    diffs_greater_than_60 = df_prices[df_prices['diffs'] >= enter_trade]
    ax.scatter(diffs_greater_than_60.index, diffs_greater_than_60[ax.get_ylabel()], color='red', zorder=2)

for ax in axs:
    diffs_less_15 = df_prices[(-exit_trade <= df_prices['diffs']) & (df_prices['diffs'] <= exit_trade)]
    ax.scatter(diffs_less_15.index, diffs_less_15[ax.get_ylabel()], color='yellow', zorder=2)

for ax in axs:
    diffs_gr_15 = df_prices[df_prices['diffs'] <= -enter_trade]
    ax.scatter(diffs_gr_15.index, diffs_gr_15[ax.get_ylabel()], color='cyan', zorder=2)

plt.xlabel('Index')
plt.title('Plot of choc, rose, strawb, and basket columns with dots where diffs >= 60 and ticks where diffs <= -60')
plt.tight_layout()
plt.show()
```
