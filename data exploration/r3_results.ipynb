```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inp_df = pd.read_csv("website_data/r3_results_prices.csv",delimiter=";")
inp_df.head(30)
```

```python
df1 = pd.DataFrame(inp_df[inp_df["product"] == "CHOCOLATE"]["mid_price"]).reset_index(drop=True)
df2 = pd.DataFrame(inp_df[inp_df["product"] == "ROSES"]["mid_price"]).reset_index(drop=True)
df3 = pd.DataFrame(inp_df[inp_df["product"] == "GIFT_BASKET"]["mid_price"]).reset_index(drop=True)
df4 = pd.DataFrame(inp_df[inp_df["product"] == "STRAWBERRIES"]["mid_price"]).reset_index(drop=True)
df_prices = pd.concat([df1,df2,df3,df4],axis=1)
df_prices = df_prices.set_axis(["choc","rose","basket","strawb"],axis=1)
df_prices
```

```python
fig,ax = plt.subplots()
ax.plot(df_prices["basket"], color = "brown")
ax2 = ax.twinx()
ax2.plot((df_prices["choc"]*4 + df_prices["rose"]*1 + df_prices["strawb"]*6 + 380))
```
